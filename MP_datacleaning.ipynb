```python
#clean up column names 
name_mapping = {
    'Route': 'name',
    'Location': 'location',
    'URL': 'url',
    'Avg Stars': 'rating',
    'Route Type': 'type',
    'Rating': 'grade',
    'Pitches': 'pitches',
    'Length': 'length',
    'Area Latitude': 'lat',
    'Area Longitude': 'lon',
    'votes': 'votes',
    'Region': 'region'
}
df = df.rename(name_mapping, axis='columns')
df.head()
```

```python
yds = df.grade.str.extract(r'5\.(\d+)', expand=False)
yds.value_counts()
```

```python
protection = df.grade.str.extract(r'(PG13|R|X)', expand=False)
protection.value_counts()
```

```python
#add grades and protection to dataframe
df['yds'] = pd.to_numeric(yds)
df['protection'] = protection
df
```

```python
#identify longest route
df[df.length == df.length.max()]
#remove longest route
df = df.drop(4748)
```

```python
#-1 rating identifies unrated routes
df[df.rating == df.rating.min()]
#fill these ratings with NA
df.loc[df['rating'] < 0, 'rating'] = np.nan
```

```python
detailed_loc = df['location'].str.split('>', expand=True)
detailed_loc = detailed_loc.rename(columns = {0:'sub_area',1:'area'})
detailed_loc
```

```python
#locations are not indexed in the same way... Just merge area and sub_area
df = df.merge(detailed_loc[['sub_area','area']], how = 'left', left_index = True, right_index = True)
```

```python
# filter out TR only routes
df = df[df.type != 'TR']
# If a route has Trad in its type then it's trad, else sport
df['type'] = np.where(df.type.str.contains('Trad'), 'Trad', 'Sport')
df.type.value_counts()
```
